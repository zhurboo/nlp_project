```python
import numpy as np
from collections import Counter
import matplotlib.pyplot as plt
```

```python
train_title_lens = []
test_title_lens = []
train_sen_lens = []
test_sen_lens = []

with open('data/title_storyline_story/train.txt', 'r') as f:
    for line in f:
        train_title_lens.append(len(line.split('<EOT>')[0].split()))
        sens = line.split('<EOL>')[-1].split('</s>')[1:]
        for sen in sens:
            train_sen_lens.append(len(sen.split()))
with open('data/title_storyline_story/valid.txt', 'r') as f:
    for line in f:
        train_title_lens.append(len(line.split('<EOT>')[0].split()))
        sens = line.split('<EOL>')[-1].split('</s>')[1:]
        for sen in sens:
            train_sen_lens.append(len(sen.split()))
with open('data/title_storyline_story/test.txt', 'r') as f:
    for line in f:
        test_title_lens.append(len(line.split('<EOT>')[0].split()))
        sens = line.split('<EOL>')[-1].split('</s>')[1:]
        for sen in sens:
            test_sen_lens.append(len(sen.split()))
```

```python
def plot(degs):
    hist = Counter(degs)
    print(len(hist))
    x,y = [],[]
    for each in hist.items():
        x.append(each[0])
        y.append(each[1]/len(degs))
    x,y = np.array(x),np.array(y)
    arg = np.argsort(y)[-50:][::-1]
    x,y = x[arg],y[arg]
    x = range(1,51,1)
    plt.plot(x,y,'.')
    plt.show()
    return set(hist.keys())
```

```python
plt.hist(train_sen_lens,100)
plt.show()
```

```python
train_d = []
test_title_d = []
test_story_d = []

with open('data/title_storyline_story/train.txt', 'r') as f:
    for line in f:
        train_d.extend(line.split('<EOT>')[0].split())
        sens = line.split('<EOL>')[-1].split('</s>')[1:]
        for sen in sens:
            train_d.extend(sen.split())
with open('data/title_storyline_story/valid.txt', 'r') as f:
    for line in f:
        train_d.extend(line.split('<EOT>')[0].split())
        sens = line.split('<EOL>')[-1].split('</s>')[1:]
        for sen in sens:
            train_d.extend(sen.split())
with open('data/title_storyline_story/test.txt', 'r') as f:
    for line in f:
        test_title_d.extend(line.split('<EOT>')[0].split())
        sens = line.split('<EOL>')[-1].split('</s>')[1:]
        for sen in sens:
            test_story_d.extend(sen.split())
```

```python
train_d = plot(train_d)
test_title_d = plot(test_title_d)
test_story_d = plot(test_story_d)
```

```python
print(len(train_d),len(test_title_d),len(test_story_d))
print(len(test_title_d-train_d),len(test_story_d-train_d))
```
